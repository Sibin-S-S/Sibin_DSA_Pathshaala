1. Create a NumPy array with values from 10 to 49 (inclusive).

2. Create two 3×3 matrices:
One filled with all 5s , One with values from 1 to 9
● Add the two matrices
● Multiply them element-wise
● Then find the transpose of the result

```python
a=np.full((3,3),5)
b=np.arange(1,10).reshape(3,3)
print(f'\nA\n = {a}\n')
print(f'\nB\n = {b}\n')

sum=np.add(a,b)
print(f"\nSum\n = {sum}\n")

prod=np.matmul(a,b)
print(f'\nMultiplication\n = {prod}')

transpose=np.transpose(prod)
print(f'\nTranspose\n = {transpose}')
```

3.Create a NumPy array of 15 random integers between 10 and 100.
● Replace all values greater than 50 with -1
● Count how many values were replaced.

```python
a=np.random.randint(10,100,size=(15))
print(a)

a1=np.where(a>50,-1,a)
print(f'Replaced = {a1}')

count=np.sum(a>50)
print(f'No of values replaced = {count}')



```

4. Generate a 5×4 array with random values between 0 and 1 (float).
● Find the row with the maximum sum of values.
● Normalize each row using the formula:

```python
a=np.random.rand(5,4)
print(f'A ={a}\n')

b=np.sum(a,axis=1)
max_row_index=np.argmax(b)
print(f"Row with maximum sum of values = {a[max_row_index]}\n")

row_max=np.max(a,axis=1,keepdims=True)
row_min=np.min(a,axis=1,keepdims=True)
normalized=(a-row_min)/(row_max-row_min)
print(f"Normalized=\n{normalized}")
```

5. arr = np.array([1, 3, 5, 3, 7, 9, 1, 5, 5])
● Find all unique elements
● Find the duplicates (elements that appear more than once)

```python
arr=np.array([1,3,5,3,7,9,1,5,5])

unique=np.unique(arr)
print(f"Unique elements : {unique}")

duplicate=np.where(np.bincount(arr)>1)[0]
print(f"Duplicate elements : {duplicate}")
```

6. Create a 6×6 NumPy array with values from 1 to 36.
● Extract the middle 4×4 submatrix.
● Then extract all even numbers from this submatrix.

```python
arr=np.arange(1,37).reshape(6,6)
print(f'A=\n{arr}\n')

new=arr[1:5,1:5]
print(f'4 x 4 =\n {new}\n')

even=new[new%2==0]
print(f'Even_numbers = {even}')
```

7. Generate an array of shape (3, 4) with values from 0 to 11.
● Reshape it to (2, 6)
● Then flatten it into a 1D array
● Convert the 1D array back to shape (4, 3)

```python
arr=np.arange(0,12).reshape(3,4)
print(f'A =\n{arr}\n')

arr1=np.reshape(arr,(2,6))
print(f'Reshaped =\n{arr1}\n')

arr2=arr1.flatten()
print(f"Flatten =\n{arr2}\n")

arr3=np.reshape(arr2,(4,3))
print(f'4 x 3\n= {arr3}')
```

8 . Create an array: a = np.array([[1, 2], [3, 4]])
● Append a new row [5, 6]
● Then append a new column [7, 8, 9]

```python
a=np.array([[1,2],[3,4]])
print(f"A =\n {a}\n")

new_row=np.array([5,6])
b=np.vstack([a,new_row])
print(f"New row =\n{b}\n")

new_column=np.array([[7],
                    [8],
                    [9]])
c=np.hstack([b,new_column])
print(f"New column =\n{c}")
```

9. a = np.array([1, 2, 3])
 b = np.array([4, 5, 6])
 c = np.array([7, 8, 9])
 ● Join them into a single array
 ● Reshape the result into a 3×3 matrix
 ● Extract the diagonal elements

```python
a=np.array([1,2,3])
b=np.array([4,5,6])
c=np.array([7,8,9])
A=np.concatenate([a,b,c])
print(f'A = {A}\n')

A1=np.reshape(A,(3,3))
print(f"3 x 3=\n{A1}\n")

A2=np.diag(A1)
print("Diagnal elements = \n",A2)
```

10. Create a NumPy array of integers from 1 to 20.
● Extract all values that are divisible by 3 or 5.
● Count how many such values exist.

```python
a=np.arange(1,21)
print(f"A\n= {a}\n")

div=a[(a%3==0) | (a%5==0)]
print(f"Divisible by 3 or 5\n= {div}\n" )

count=div.size
print(f"No of values = {count}")
```
