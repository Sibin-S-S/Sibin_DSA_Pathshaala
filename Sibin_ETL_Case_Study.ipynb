```python
import pandas as pd
import numpy as np
```

 Load the dataset into a DataFrame

```python
data=pd.read_csv("/content/all_month.csv")
df=pd.DataFrame(data)
df
```

Identify number of records and unique locations

```python
print(f"Number of rows : {len(df)}")
print(f"Unique locations : {df['place'].unique()}")
```

 Print top 5 rows and column names

```python
df.head()
```

```python
print(data.columns)
```

Convert time to datetime


```python
df['time']=pd.to_datetime(df['time'])
df['time'].dtype
```

 Drop records with missing latitude, longitude, mag


```python
df=df.dropna(subset=['latitude','longitude','mag'])
```

Filter only earthquakes with magnitude >= 4.0

```python
df[df['mag']>=4.0]
```

 Add a column day_of_week from time

```python
df['day_of_week']=df['time'].dt.day_name()
df
```

Create a column severity_level based on magnitude:
○ <4.0: "Low"
○ 4.0 - 6.0: "Moderate"
○ 6.0+: "High"


```python
df['Severity_Level']=np.where(df['mag']<4.0,'Low',np.where((df['mag']>=4) & (df['mag']<=6),'Moderate','High'))
df
```

Count number of earthquakes per place

```python
df.groupby('place').size()
```

 Compute average magnitude and max depth per day


```python
df.groupby(df['time'].dt.date).agg(avg_magnitude=('mag','mean'),max_depth=('depth','max'))
```

● Save:
○ Cleaned dataset to cleaned_earthquakes.csv
○ Summary data to earthquake_summary.csv

```python
df.to_csv('Cleaned_earthquakes.csv',index=False)
```

```python
df1=pd.read_csv('Cleaned_earthquakes.csv')
import pandas as pd


df1['time'] = pd.to_datetime(df1['time'], format='mixed')

summary = df1.groupby(df1['time'].dt.date).agg(
    avg_magnitude=('mag', 'mean'),
    max_depth=('depth', 'max')
).reset_index()

summary.to_csv("earthquake_summary.csv", index=False)

print(summary.head())
```

Store into SQLite table

```python
import sqlite3
conn=sqlite3.connect('eathquakes.db')
df.to_sql("Earthquakes", conn, if_exists="replace", index=False)
summary.to_sql('Earthquake_summary',conn,if_exists='replace',index=False)
tables = pd.read_sql("SELECT name FROM sqlite_master WHERE type='table';", conn)
print(tables)
conn.close()
```
