```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# List of columns you want to process
cols = [
    'Sale Price',
    'No of Bedrooms',
    'Flat Area (in Sqft)',
    'Lot Area (in Sqft)',
    'Area of the House from Basement (in Sqft)',
    'Basement Area (in Sqft)',
    'Age of House (in Years)',
    'Living Area after Renovation (in Sqft)',
    'Lot Area after Renovation (in Sqft)'
]

for col in cols:
    # 1. IQR bounds
    q1 = np.quantile(df[col], 0.25)
    q2 = np.quantile(df[col], 0.75)
    iqr = q2 - q1
    upb = q2 + (1.5 * iqr)
    lwb = q1 - (1.5 * iqr)

    print(f"\nColumn: {col}")
    print("Higher bound:", upb)
    print("Lower bound:", lwb)
    print("IQR:", iqr)

    # 2. Collect outliers
    outliers = [x for x in df[col] if x > upb or x < lwb]
    print("No. of outliers:", len(outliers))

    # 3. Clip values
    df[col] = df[col].clip(lower=lwb, upper=upb)

    plt.figure(figsize=(5,6))
    plt.boxplot(df[col])
    plt.title(col)
    plt.show()

```
