```python
import pandas as pd
populations = pd.read_csv('/content/population.csv')
populations
```

```python
import pandas as pd
gdp1 = pd.read_excel('/content/gdp.xlsx')
gdp1
```

```python
import pandas as pd
internet_user = pd.read_json('/content/internet_users.json')
internet_user
```

```python
literacy_rates = pd.read_xml('/content/literacy_rate.xml')
literacy_rates
```

```python
# Standardize country names
populations = populations.rename(columns={'country': 'Country'})
internet_user = internet_user.rename(columns={'country': 'Country'})
literacy_rates = literacy_rates.rename(columns={'name': 'Country'})
```

```python
# Convert numeric columns to appropriate types
```

```python
for i,df in enumerate([populations,gdp1,internet_user,literacy_rates], start =1):
  print(f'Dataset {i} has nulls? {df.isnull().values.any()}')
```

```python
# Create new columns such as internet penetration rate (internet users / population  100)
internet_user = pd.merge(
     internet_user,
     populations[['Country', 'population']],
     on='Country',
     how='left'
 )
internet_user['internet_users_percentage'] = (
     internet_user['internet_users'] / internet_user['population'] * 100
 )
internet_user = internet_user[['Country', 'internet_users', 'internet_users_percentage']]
internet_user
```

### Integrate

```python
#  Merge the datasets based on the country name

mergdata = pd.merge(
     populations,
     gdp1,
     on='Country',
     how='left'
 )
mergdata = pd.merge(
     mergdata,
     internet_user,
     on='Country',
     how='left'
 )
mergdata = pd.merge(
     mergdata,
     literacy_rates,
     on='Country',
     how='left'
 )
mergdata
```

```python
#Handle unmatched or missing records
mergdata.isnull().values.any()
```

```python
valid_rep = {
    'Missing values': mergdata.isnull().to_dict(),
    'Duplicate rows': mergdata.duplicated().sum(),
    'Invalid Populations': (mergdata['population'] <= 0).sum(),
    'Invalid GDP': (mergdata['GDP'] < 0).sum(),
    'Invalid Internet Users': (mergdata['internet_users'] < 0).sum(),
}
valid_rep
```

```python
# Find countries with highest internet penetration rates
top_int_pen = mergdata.nlargest(5, 'internet_users_percentage')
top_int_pen
```

```python
# Calculate average literacy rate across countries
avg_literacy = mergdata['literacy_rate'].mean()
avg_literacy.item()
```

```python
# Investigate relationships between GDP, population, literacy, and internet usage
gdp_corr = mergdata['GDP'].corr(mergdata['population'])
literacy_corr = mergdata['literacy_rate'].corr(mergdata['internet_users_percentage'])
print(f'The correlation between GDP and population is  ', gdp_corr)
print(f'The correlation between literacy and internet usage is  ', literacy_corr)
```

```python
# Create an SQLite database
import sqlite3
conn = sqlite3.connect('mergdata.db')
cursor = conn.cursor()
# Load the final merged dataset into a database table
mergdata.to_sql('mergdata', conn, if_exists='replace', index=False)
```

```python
cursor.execute('select * from mergdata')
```

```python
cursor.fetchall()
```

```python
# Write and execute SQL queries to extract insights
# a. Countries with highest internet penetration
cursor.execute('select Country, internet_users_percentage from mergdata order by internet_users_percentage  desc limit 5')
cursor.fetchall()
```

```python
# Average literacy rate
cursor.execute('select avg(literacy_rate) from mergdata')
cursor.fetchall()
```

```python
# Correlation between literacy rate and internet penetration
correlation = mergdata['literacy_rate'].corr(mergdata['internet_users_percentage'])
print(f'The correlation is  ', correlation)
```

```python
# GDP per capita analysis
cursor.execute('select Country,gdp,population,(gdp * 1.0/population) AS gdp_per_capita  from mergdata order by gdp_per_capita desc limit 5')
cursor.fetchall()
```

```python
# Which countries have a GDP per capita above $10,000?
cursor.execute('select Country,gdp,population,(gdp * 1.0/population) AS gdp_per_capita  from mergdata where gdp_per_capita > 10000  order by gdp_per_capita desc')
cursor.fetchall()
```

```python
# What is the total population covered in the dataset?
cursor.execute('select sum(population) from mergdata')
cursor.fetchall()
```

```python
# Which countries have the lowest literacy rates
cursor.execute('select Country,literacy_rate ,internet_users_percentage from mergdata order by literacy_rate asc ')
cursor.fetchall()
```

```python
# What are the top 5 wealthiest countries by total GDP, and how does that compare with population size?
cursor.execute('select Country,gdp,population,(gdp * 1.0/population) AS gdp_per_capita  from mergdata order by gdp desc limit 5')
cursor.fetchall()
```

```python
# Find countries where internet users exceed 70% of the population
cursor.execute('select Country,population,internet_users,internet_users_percentage from mergdata where internet_users_percentage > 70')
cursor.fetchall()
```

```python
# What is the average GDP per capita for countries with internet penetration above 50%?
cursor.execute('select avg(gdp * 1.0 /population) AS avg_gdpc from mergdata where (internet_users * 100 /population) > 50;')
cursor.fetchall()
```

```python
# How many countries have a literacy rate above 90%, and what is their average internet penetration?
cursor.execute('select count(*) AS num_con,AVG(internet_users * 100.0 / population) AS avg_intp from mergdata where literacy_rate > 90; ')
cursor.fetchall()
```
