```python
def grocery_store_billing():
    print("\n1. Grocery Store Billing")
    total = 0
    while True:
        item = input("Enter item name (or 'done' to finish): ")
        if item.lower() == 'done':
            break
        try:
            price = float(input(f"Enter price of {item}: ₹"))
            total += price
        except:
            print("Invalid price.")
    if total > 500:
        total *= 0.9
    print(f"Total Bill: ₹{total:.2f}")


def exam_result_calculator():
    print("\n2. Exam Result Calculator")
    marks = []
    for i in range(1, 6):
        marks.append(float(input(f"Enter marks for subject {i}: ")))
    percentage = sum(marks) / 5
    if percentage >= 90:
        grade = 'A'
    elif percentage >= 75:
        grade = 'B'
    elif percentage >= 50:
        grade = 'C'
    else:
        grade = 'Fail'
    print(f"Percentage: {percentage:.2f}% | Grade: {grade}")


def atm_simulation():
    print("\n3. ATM Simulation")
    balance = 1000
    while True:
        print("\nOptions: 1.Deposit 2.Withdraw 3.Check Balance 4.Exit")
        choice = input("Enter choice (1/2/3/4): ")
        if choice == '1':
            amount = float(input("Enter amount to deposit: ₹"))
            balance += amount
            print(f"Deposited ₹{amount:.2f}. New balance: ₹{balance:.2f}")
        elif choice == '2':
            amount = float(input("Enter amount to withdraw: ₹"))
            if amount <= balance:
                balance -= amount
                print(f"Withdrawn ₹{amount:.2f}. New balance: ₹{balance:.2f}")
            else:
                print("Insufficient funds.")
        elif choice == '3':
            print(f"Current balance: ₹{balance:.2f}")
        elif choice == '4':
            break
        else:
            print("Invalid choice.")


def cricket_score_tracker():
    print("\n4. Cricket Score Tracker")
    scores = {}
    for i in range(1, 6):
        name = input(f"Enter player {i} name: ")
        score = int(input(f"Enter score for {name}: "))
        scores[name] = score
    top_player = max(scores, key=scores.get)
    print(f"Top Scorer: {top_player} with {scores[top_player]} runs.")
    update_name = input("Enter player name to update score: ")
    if update_name in scores:
        new_score = int(input(f"Enter new score for {update_name}: "))
        scores[update_name] = new_score
        print("Score updated.")


def restaurant_order_system():
    print("\n5. Restaurant Order System")
    menu = {'Pizza': 250, 'Burger': 150, 'Pasta': 200, 'Sandwich': 100}
    print("Menu:", menu)
    total = 0
    orders = []
    while True:
        item = input("Order item (or 'done' to finish): ")
        if item.lower() == 'done':
            break
        if item in menu:
            total += menu[item]
            orders.append(item)
        else:
            print("Item not in menu.")
    if 'Pizza' in orders:
        total -= 50
    print(f"Final Bill: ₹{total:.2f}")


def temperature_conversion_tool():
    print("\n6. Temperature Conversion Tool")
    celsius = float(input("Enter temperature in Celsius: "))
    fahrenheit = (celsius * 9/5) + 32
    print(f"{celsius:.2f}°C = {fahrenheit:.2f}°F")


def online_shopping_cart():
    print("\n7. Online Shopping Cart")
    cart = []
    while True:
        action = input("Add, Remove, View, or 'checkout' to finish: ").lower()
        if action == 'add':
            item = input("Enter item to add: ")
            cart.append(item)
        elif action == 'remove':
            item = input("Enter item to remove: ")
            if item in cart:
                cart.remove(item)
            else:
                print("Item not in cart.")
        elif action == 'view':
            print("Cart:", cart)
        elif action == 'checkout':
            print("Final Cart:", cart)
            break
        else:
            print("Invalid action.")


def student_attendance_tracker():
    print("\n8. Student Attendance Tracker")
    students = []
    present, absent = 0, 0
    n = int(input("Number of students: "))
    for _ in range(n):
        name = input("Enter student name: ")
        status = input("P (Present) or A (Absent): ").upper()
        students.append((name, status))
        if status == 'P':
            present += 1
        else:
            absent += 1
    print(f"Present: {present} | Absent: {absent}")


def cinema_ticket_booking_system():
    print("\n9. Cinema Ticket Booking System")
    seats = 50
    while True:
        print(f"Available seats: {seats}")
        tickets = int(input("How many tickets to book (0 to exit): "))
        if tickets == 0:
            break
        if tickets <= seats:
            seats -= tickets
            print(f"Booking confirmed for {tickets} ticket(s).")
        else:
            print("Not enough seats available.")


def bank_loan_eligibility_checker():
    print("\n10. Bank Loan Eligibility Checker")
    salary = float(input("Enter salary: ₹"))
    credit_score = int(input("Enter credit score: "))
    age = int(input("Enter age: "))
    if salary >= 25000 and credit_score >= 650 and 21 <= age <= 60:
        print("Eligible for loan.")
    else:
        print("Not eligible for loan.")


# Execute all functions
grocery_store_billing()
exam_result_calculator()
atm_simulation()
cricket_score_tracker()
restaurant_order_system()
temperature_conversion_tool()
online_shopping_cart()
student_attendance_tracker()
cinema_ticket_booking_system()
bank_loan_eligibility_checker()
```
