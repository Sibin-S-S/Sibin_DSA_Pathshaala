The join used here is left join.Dataset is treated as the main table containing the full list of titles.

A left join keeps all rows from titles, regardless of whether there is matching data in the other sheets.

This avoids losing titles that have missing director/country/cast/category info, which would happen with an inner join.

```python


file_path = "netflix.xlsx"


titles = pd.read_excel('/content/netflix.xlsx', sheet_name="netflix_titles")
directors = pd.read_excel('/content/netflix.xlsx', sheet_name="netflix_titles_directors")
countries = pd.read_excel('/content/netflix.xlsx', sheet_name="netflix_titles_countries")
cast = pd.read_excel('/content/netflix.xlsx', sheet_name="netflix_titles_cast")
category = pd.read_excel('/content/netflix.xlsx', sheet_name="netflix_titles_category")

master_df = (
    titles
    .merge(directors, on="show_id", how="left")
    .merge(countries, on="show_id", how="left")
    .merge(cast, on="show_id", how="left")
    .merge(category, on="show_id", how="left")
)

master_df.head()
```

The join used here is left join.Dataset is treated as the main table containing the full list of titles.

A left join keeps all rows from titles, regardless of whether there is matching data in the other sheets.

This avoids losing titles that have missing director/country/cast/category info, which would happen with an inner join.

### Data Cleaning

```python
import pandas as pd
import numpy as np


df1 = master_df.copy()

df1["show_id"] = df1["show_id"].astype("Int64")
df1["release_year"] = df1["release_year"].astype("Int64")
df1["duration_minutes"] = pd.to_numeric(df["duration_minutes"], errors="coerce")
df1["duration_seasons"] = pd.to_numeric(df["duration_seasons"], errors="coerce")
df1["date_added"] = pd.to_datetime(df1["date_added"], errors="coerce")


df1["year_added"] = df1["date_added"].dt.year.astype("Int64")
df1["type"] = df1["type"].fillna("Unknown Type")
df1["title"] = df1["title"].fillna("Untitled")
df1["rating"] = df1["rating"].fillna("Unknown Rating")


df1["description"] = df1["description"].fillna("No description available")


df1["director"] = df1["director"].fillna("Unknown Director")
df1["country"] = df1["country"].fillna("Unknown Country")
#  For null value in country column I did not drop these data,inplace i replaced with country unknown since there will be no data loss and helps in better accuracy"
df1["cast"] = df1["cast"].fillna("Unknown Cast")
df1["listed_in"] = df1["listed_in"].fillna("Uncategorized")




df1["duration_minutes"] = pd.to_numeric(df1["duration_minutes"], errors="coerce")
df1["duration_seasons"] = pd.to_numeric(df1["duration_seasons"], errors="coerce")


median_duration = df1["duration_minutes"].median(skipna=True)


df1["duration_minutes"] = df1["duration_minutes"].fillna(median_duration)


df1 = df1.drop(columns=["duration_seasons"])


master_df[["duration_minutes"]].info()
master_df["duration_minutes"].isna().sum()




print(df1.isna().sum())
```

### EDA

```python
import matplotlib.pyplot as plt


df1_exploded_cast = df1.copy()
df1_exploded_cast["cast"] = (
    df1_exploded_cast["cast"]
    .str.split(",")
    .apply(lambda lst: [x.strip() for x in lst] if isinstance(lst, list) else lst)
)
df1_exploded_cast = df1_exploded_cast.explode("cast").reset_index(drop=True)

top_actors = df1_exploded_cast["cast"].value_counts().head(10)

plt.figure(figsize=(10, 6))
top_actors.sort_values(ascending=True).plot(kind="barh", color="coral")
plt.xlabel("Number of Titles")
plt.ylabel("Actor")
plt.title("Top 10 Most Frequent Actors on Netflix")
plt.tight_layout()
plt.show()
```

Based on the visualisation,the insights that i get are that Alfred Molina has acted in most titles.And it is dominated by Unknown Casts this tells you that a very large share of titles are missing proper cast information in the dataset.
