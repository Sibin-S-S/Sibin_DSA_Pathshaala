1. Import pandas , numpy

```python
import numpy as np
import pandas as pd
```

2. Load dataset using pd.read_csv()

```python
url = 'https://raw.githubusercontent.com/laxmimerit/All-CSV-ML-Data-Files-Download/master/IMDB-Movie-Data.csv'
data =pd.read_csv(url)
data
```

3) Display: shape, head, column names, datatypes, missing value

```python
print(f'Shape = {data.shape}\n')
print(f'Head = {data.head()}\n')
print(f'Column Names = {data.columns}\n')
print(f'Datatypes = {data.dtypes}\n')
print(f'Missing values ={data.isna().sum()}')
```

4) Check and remove duplicates

```python
num_dupes_before = data.duplicated().sum()
print(f"\nDuplicates before removal: {num_dupes_before}")
df = data.drop_duplicates()
print(f"Duplicates after removal: {df.duplicated().sum()}")
```

5) Rename columns ("Title" > "Movie_Title")

```python
data.rename(columns={'Title':'Movie_Title'})
```

6) Get unique genres


```python
data['Genre'].unique()
```

7) Extract number of genres per movie

```python
data.groupby('Genre').size()
```

8) Handle missing values in 'Revenue (Millions)' and 'Metascore


```python
data = data.dropna(subset=['Revenue (Millions)', 'Metascore'])
data
```

9) Top 5 movies by Rating

```python
data.sort_values(by='Rating', ascending=False).head(5)
```

10) Movies with highest/lowest revenue

```python
Highest_Revenue = data.sort_values(by='Revenue (Millions)', ascending=False).head(1)
Lowest_Revenue = data.sort_values(by='Revenue (Millions)', ascending=True).head(1)
print(f'Highest Revenue = {Highest_Revenue}')
print(f'Lowest Revenue = {Lowest_Revenue}')
```

11) Avg. rating by Director

```python
data.groupby('Director')['Rating'].mean()
```

12) Movies per year

```python
data['Year'].value_counts()
```

13) Movies with rating > 8 & revenue > 100M


```python
data[(data['Rating'] > 8) & (data['Revenue (Millions)'] > 100)]
```

14) Group by Director: avg Rating, Revenue

```python
print("\nAverage Rating and Revenue by Director:")
print(data.groupby('Director')[['Rating','Revenue (Millions)']].mean().head())
```

15) Group by Year: count, avg Rating

```python
data.groupby('Year').agg({'Title':'count','Rating':'mean'})
```

16) Create a new column Profitability:
a) Revenue divided by Duration
b) Find top 10 most profitable movies

```python
data['Profitability'] = data['Revenue (Millions)']/data['Runtime (Minutes)']
data.sort_values(by='Profitability', ascending=False).head()
```

17) Create a column that categorizes movies by rating:
a) Rating < 6: Poor
b) 6 â‰¤ Rating < 7.5: Average
c) Rating â‰¥ 7.5: Excellent

```python
import numpy as np
data['Rating(Category)'] = np.where(data['Rating'] < 6, 'Poor', np.where(data['Rating'] < 7.5, 'Average', 'Excellent'))
print(data[['Title','Rating' , 'Rating(Category)']])
```

18) Create a custom function to score movies based on:
i) Apply row['Rating'] * 10 + row['Revenue (Millions)'] as High score
Apply a function to dataset to assign the score

```python
def high_score(row):
    return row['Rating'] * 10 + row['Revenue (Millions)']
data['High_Score'] = data.apply(high_score, axis=1)
print(data[['Title', 'Rating', 'Revenue (Millions)', 'High_Score']])
```
